```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("LLM_vordlus.csv")
data['Aasta']= pd.to_datetime(data['Aasta'])
```

```python
data
```

```python
plt.figure(figsize=(8,4))
plt.scatter(data.Aasta, data.Parameetreid, s=data.Korpus_suurus, alpha=0.8, c = "#4DA4EA")
mlabels = ["10^9", "10^10", "10^11", "10^12"]
msizes = [20, 100, 200, 500]
markers = []
for size in msizes:
    markers.append(plt.scatter([],[], s=size, label=size, c = "#4DA4EA", alpha = 0.8))
plt.legend(handles=markers, loc = 'lower right', labels = mlabels, title = "Treeningandmete suurus", labelspacing = 1.9, 
          bbox_to_anchor = (1.35, 0), borderpad = 1)

plt.ylabel("Parameetrite arv")
plt.xlabel("Mudeli avaldamise kuupäev")
plt.xticks(fontsize=9)
for i, txt in enumerate(data.Nimi):
    if txt == "Chinchilla":
        plt.annotate(txt, (data.Aasta[i], data.Parameetreid[i]), textcoords='offset points', xytext = (-15,-20))
    elif txt == "MT-NLG":
        plt.annotate(txt, (data.Aasta[i], data.Parameetreid[i]), textcoords='offset points', xytext = (-40,10))
    elif txt in ["LLaMa", "GLaM"]:
        plt.annotate(txt, (data.Aasta[i], data.Parameetreid[i]), textcoords='offset points', xytext = (-15,15))
    elif txt == "RoBERTa":
        plt.annotate(txt, (data.Aasta[i], data.Parameetreid[i]), textcoords='offset points', xytext = (-25,10))
    else:
        plt.annotate(txt, (data.Aasta[i], data.Parameetreid[i]), textcoords='offset points', xytext = (-15,10))
plt.yscale('log')
plt.ylim(top=4000000000000)
plt.savefig('LLM_mudelid.JPG')
plt.show()
```
